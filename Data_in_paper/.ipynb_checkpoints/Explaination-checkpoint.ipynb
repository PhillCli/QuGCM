## ${\bf H}$ and ${\bf S}$ Matrices for H4 models ($\alpha = 0.005$ and $\alpha = 0.500$)

Hamiltonian matrices $H$ for both H4 models are saved in `QuGCM/Example_Data/`

* `QuGCM/Example_Data/h4_linear_ham.txt` is for $\alpha = 0.500$

* `QuGCM/Example_Data/h4_square_ham.txt` is for $\alpha = 0.005$

(For all files saved in `QuGCM/Example_Data/`,  "L" or "linear" refers to $\alpha = 0.500$ model, and "S" or "square" refers to $\alpha = 0.005$ model)

The orbital conventsion is shown below, which is different from Qiskit. We provide the code for reading both Hamiltonian matrices and form Qiskit `FerionicOp` object in `QuGCM/src/example_molecules.py`. People may check `QuGCM/Demo.ipynb` to see the specific usage.



### Generation of ${\bf H}$ and ${\bf S}$ from molecular Hamiltonian matrices

The code is exactly shown in `QuGCM/Demo.ipynb`, while the computation results in paper are with 15 basis vectors. For the choices of $\{t\}$ of both models, they are saved in 

* `QuGCM/Data_in_paper/Exp15B_EXPQWC_H4L_Alt15B_max100t_t.npy` for $\alpha = 0.500$

* `QuGCM/Data_in_paper/Exp15B_EXPQWC_H4S_Alt15B_max100t_t.npy` for $\alpha = 0.005$

To obtain ${\bf H}$ and ${\bf S}$ for $\alpha = 0.500$ model, we can

```python
import numpy as np

H_l = np.load("Exp15B_EXPQWC_H4L_Alt15B_max100t_H_qwc.npy")
S_l = np.load("Exp15B_EXPQWC_H4L_Alt15B_max100t_S_qwc.npy")
```

We can plot both matrices

```python
import sys
sys.path.insert(1, '../src')
from qugcm import plot_sh
basis_displayed = ['R0','+R1', '-R1', '+R2', '-R2', 
                    '+R3', '-R3', '+R4', '-R4','R7','R8',
                   '+R2+R1', '+R2-R1', '-R2+R1', '-R2-R1'] 
plot_sh(S_l, H_l, basis_displayed, save_name=None, exp_name=r'$\alpha = 0.500$')
```

Similarily, for $\alpha = 0.005$ model

```python
H_s = np.load("Exp15B_EXPQWC_H4S_Alt15B_max100t_H_qwc.npy") # by just chaing "H4L" to "H4S" in the file name
S_s = np.load("Exp15B_EXPQWC_H4S_Alt15B_max100t_S_qwc.npy") # by just chaing "H4L" to "H4S" in the file name
plot_sh(S_s, H_s, basis_displayed, save_name=None, exp_name=r'$\alpha = 0.005$')
```

### Usage

With provided ${\bf H}$ and ${\bf S}$ above, we can directly solve for lowest energies in both models by solving generalize eigenvalue problems.

For $\alpha = 0.500$, 

```python
import scipy.linalg as sl
w_l, v_l = sl.eigh(H_l, S_l)
w_l[0]
```

For $\alpha = 0.005$, 

```python
w_s, v_s = sl.eigh(H_s, S_s)
w_s[0]
```

## Influence of finite measurements

In Eq.(38) and (39), we show the upper bounds of standard deviation of each matrix entry. The numerical value of **numerators** of those bounds (i.e., $\sum_{G_{{\bf H}_{pq}}}\sqrt{\sum_{i,j,\in G_{{\bf H}_{pq}}} |h_i h_j|}$ and $\sum_{G_{{\bf S}_{pq}}}\sqrt{\sum_{i,j,\in G_{{\bf S}_{pq}}} |h_i h_j|}$) are saved in 

* `QuGCM/Data_in_paper/H4L_H_maxvar.npy`  for ${\bf H}$ in $\alpha = 0.500$ model 

* `QuGCM/Data_in_paper/H4L_S_maxvar.npy`  for ${\bf S}$ in $\alpha = 0.500$ model

* `QuGCM/Data_in_paper/H4S_H_maxvar.npy`  for ${\bf H}$ in $\alpha = 0.005$ model

* `QuGCM/Data_in_paper/H4S_S_maxvar.npy`  for ${\bf S}$ in $\alpha = 0.005$ model

For example, in $\alpha = 0.500$ model, the numerators of upper bound for $\epsilon_{{\bf H}_{2,3}}$ (assuming indexing from 0) is

```python
np.load('H4L_H_maxvar.npy')[2,3]
```

So, if the number of measurements is 1000, then the upper bound should be

```python
np.load('H4L_H_maxvar.npy')[2,3]/np.sqrt(1000)
```

Those seeds are random prime numbers generated by `randprime()` in `sympy `. One seed corresponds to one set of $t_i$. For example, to generate a set with in range $[0,1)$, we do 

```python
from numpy.random import Generator, PCG64

t_seed = 34206373 # select a seed
min_t = 0 # select the lower bound
max_t = 1 # select the upper bound

rng = Generator(PCG64(t_seed))
ts = min_t + rng.random(6)*max_t # only non-composite, [min_t, max_t)
ts = np.append([1], ts) # Inserting 1 for R0
ts
```
